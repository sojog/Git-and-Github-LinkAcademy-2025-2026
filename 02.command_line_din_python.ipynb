Creați pe Desktop (doar din linie de comandă) un folder cu numele terminal_test
În folderul terminal_test creați un fișier cu numele print_txt.py, un folder cu numele print_test 
În fișierul print_txt.py introduceți textul 
"print('Hello World')"
Mutați fișierul fișierul print_txt.py  din folderul terminal_test în folderul print_test
Rulați în linie de comandă fișierul

```python
import subprocess
```

```python
create_folder_cmd = "mkdir terminal_test"
```

```python
subprocess.call(create_folder_cmd, shell=True)
```

```python
create_file_in_folder = "echo > terminal_test/print_txt.py"
subprocess.call(create_file_in_folder, shell=True)
```

```python
create_file_in_folder = """echo "print('Hello World')" > terminal_test/print_txt.py"""
subprocess.call(create_file_in_folder, shell=True)
```

```python
create_another_folder = "mkdir terminal_test/print_test"
subprocess.call(create_another_folder, shell=True)
```

```python
move_file_into_folder = "mv terminal_test/print_txt.py terminal_test/print_test/print_txt.py"
subprocess.call(move_file_into_folder, shell=True)
```
